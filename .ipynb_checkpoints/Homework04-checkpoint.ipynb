```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_file = "Resources/schools_complete.csv"
student_file = "Resources/students_complete.csv"


# Read School and Student Data File and store into Pandas Data Frames
school_datadf = pd.read_csv(school_file)
student_datadf = pd.read_csv(student_file)


# Combine the data into a single dataset
schoolstud_comp = pd.merge(student_datadf, school_datadf, how="left", on=["school_name", "school_name"])
```

```python
total_school_names = school_datadf["school_name"].count()
total_school_names
```

```python
total_students = student_datadf["student_name"].count()
total_students
```

```python
total_budget = school_datadf["budget"].sum() 
total_budget
```

```python
avg_math_score = student_datadf["math_score"].mean()
avg_math_score
```

```python
avg_reading_score = student_datadf["reading_score"].mean()
avg_reading_score
```

```python
#Percent of Students Passing Math 
stud_passmath = student_datadf.loc[(student_datadf)["math_score"] >= 70]
stud_passmath

amnt_passmath = stud_passmath["math_score"].count()
amnt_passmath 

percent_passmath = (amnt_passmath)/(total_students) *100
percent_passmath
```

```python
#Percent of Students Passing Reading 
stud_passread = student_datadf.loc[(student_datadf)["reading_score"] >=70]
stud_passread 

amnt_passread = stud_passread["reading_score"].count()
amnt_passread

percent_passread = (amnt_passread)/(total_students) *100
percent_passread
```

```python
#Overall Passing Rate 
overall_pass_perc = ((percent_passmath) + (percent_passread))/2
overall_pass_perc
```

```python
#Overview of District Summary
District_summary = pd.DataFrame({ 

"Total Schools":[total_school_names],
"Total Students":[total_students],
"Total Budget":[total_budget],
"Average Math Score":[avg_math_score],
"Average Reading Score":[avg_reading_score],
"Percent(%)Passing Math":[percent_passmath],
"Percent(%)Passing Reading":[percent_passread],
"Overall Passing Rate(%)":[overall_pass_perc]})


District_summary
```
